## Bank Marketing Dataset

Abstract: The data is related with direct marketing campaigns (phone calls) of a Portuguese banking institution. The classification goal is to predict if the client will subscribe a term deposit (variable y).

There are four datasets:
1) bank-additional-full.csv with all examples (41188) and 20 inputs, ordered by date (from May 2008 to November 2010), very close to the data analyzed in [Moro et al., 2014]\
2) bank-additional.csv with 10% of the examples (4119), randomly selected from 1), and 20 inputs.\
3) bank-full.csv with all examples and 17 inputs, ordered by date (older version of this dataset with less inputs).\
4) bank.csv with 10% of the examples and 17 inputs, randomly selected from 3 (older version of this dataset with less inputs).\
The smallest datasets are provided to test more computationally demanding machine learning algorithms (e.g., SVM).

The classification goal is to predict if the client will subscribe (yes/no) a term deposit (variable y).

Attribute Information:

### Input variables:
##### bank client data:
1 - age (numeric)\
2 - job : type of job (categorical: 'admin.','blue-collar','entrepreneur','housemaid','management','retired','self-employed','services','student','technician','unemployed','unknown')\
3 - marital : marital status (categorical: 'divorced','married','single','unknown'; note: 'divorced' means divorced or widowed)<br>
4 - education (categorical: 'basic.4y','basic.6y','basic.9y','high.school','illiterate','professional.course','university.degree','unknown')<br>
5 - default: has credit in default? (categorical: 'no','yes','unknown')<br>
6 - housing: has housing loan? (categorical: 'no','yes','unknown')<br>
7 - loan: has personal loan? (categorical: 'no','yes','unknown')<br>

##### related with the last contact of the current campaign:
8 - contact: contact communication type (categorical: 'cellular','telephone')<br>
9 - month: last contact month of year (categorical: 'jan', 'feb', 'mar', ..., 'nov', 'dec')<br>
10 - day_of_week: last contact day of the week (categorical: 'mon','tue','wed','thu','fri')<br>
11 - duration: last contact duration, in seconds (numeric). Important note: this attribute highly affects the output target (e.g., if duration=0 then y='no'). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.<br>

##### other attributes:
12 - campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)<br>
13 - pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric; 999 means client was not previously contacted)<br>
14 - previous: number of contacts performed before this campaign and for this client (numeric)<br>
15 - poutcome: outcome of the previous marketing campaign (categorical: 'failure','nonexistent','success')<br>

##### social and economic context attributes
16 - emp.var.rate: employment variation rate - quarterly indicator (numeric)<br>
17 - cons.price.idx: consumer price index - monthly indicator (numeric)<br>
18 - cons.conf.idx: consumer confidence index - monthly indicator (numeric)<br>
19 - euribor3m: euribor 3 month rate - daily indicator (numeric)<br>
20 - nr.employed: number of employees - quarterly indicator (numeric)<br>

Output variable (desired target):<br>
21 - y - has the client subscribed a term deposit? (binary: 'yes','no')<br>

# I. Exploring Dataset with Pandas

## I.1 Data Preparation

- Get the 'Bank Marketing (with social/economic context)' data from [here](https://archive.ics.uci.edu/ml/datasets/Bank+Marketing).
- The dataset used is the bank-additional-full.csv which includes the most complete data points and features.


```python
# import numerical processing packages
import numpy as np
import pandas as pd

# import visualization packages
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use(['ggplot'])

# import machine learning algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# import other required packages for creating ML models
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# import visualization packages for decision tree
import sklearn.tree as tree
import pydotplus
from six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import graphviz
```

```python
# import csv, note that the csv has semicolon delimiter

df = pd.read_csv("dataset/bank-additional-full.csv", 
                 sep=';')
df
```

```python
# check for any null values

df.isnull().sum()
```

```python
# note that in the dataset website (see notebook header for description), it mentions that there are some rows that have 'unknown' values
# check how many data have 'unknown' values

len(df[df['default'].str.contains('unknown')])
```

```python
# we will now treat the 'unknown' value as NaN, and then drop them
# this was done because even after we remove all values that have 'unknown' values, we will still have >30,000 data out of the initial total >40,000 data
# we assume that with this amount of data, we will still be able to get good results for the ML models

df_clean = pd.read_csv("dataset/bank-additional-full.csv", 
                        sep=';', 
                        na_values="unknown")
df_clean = df_clean.dropna()
df_clean
```

```python
# we will now check if there are any duplicated data

# see how many rows are within the original dataframe
print("Number of rows before duplicates removal:", len(df_clean))

# check for duplicates and how many of them are present
duplicate_data = df_clean[df_clean.duplicated()]
print("Number of duplicated rows:", len(duplicate_data))

# drop all the duplicated data and see how many rows remain to check if they've been dropped successfully
df_clean.drop_duplicates(inplace=True)
print("Number of rows after duplicates removal:", len(df_clean))
```

```python
# check the name of the columns
# please read notebook header for descriptions of each column

df_clean.columns
```

```python
# according to the description of the dataset (see header at the beginning of this notebook), there are value of '999' which indicate that client was not previously contacted
# therefore '999' is not reflecting actual conditions unlike the other rows from the same column
# check how many are there to see if we can drop some of the rows or not

df_clean['pdays'].value_counts()
```

- In the dataset description, it was mentioned that: _"Using the rminer package and R tool (http://cran.r-project.org/web/packages/rminer/), we found that the addition of the five new social and economic attributes (made available here) lead to substantial improvement in the prediction of a success, even when the duration of the call is not included. Note: the file can be read in R using: d=read.table("bank-additional-full.csv",header=TRUE,sep=";")."_ Therefore, we will drop 'duration' column.

- 'day_of_week' (_last contact day of the week (categorical: "mon","tue","wed","thu","fri")_) will also be considered irrelevant since it is assumed to not have any logical relationship with whether or not someone is subscribed a term deposit.

- 'month' (_last contact month of year (categorical: "jan", "feb", "mar", ..., "nov", "dec")_) will also be considered irrelevant since in this dataset, we do not know whether these data points were taken from a particular year or more than a year. We also do not know when the subscription did occur for those who was subscribed the term deposit. Therefore, the timeframe is not relevant.

- 'pdays' (_number of days that passed by after the client was last contacted from a previous campaign (numeric; 999 means client was not previously contacted)_) will also be considered irrelevant since there are 29,168 rows that have "999" value, meaning the majority of the customers was not previously contacted, and rather than dropping the rows with '999' value, it is better to drop the column.

```python
# drop the 4 columns mentioned above

df_clean.drop(columns=['duration'], inplace=True)
df_clean.drop(columns=['day_of_week'], inplace=True)
df_clean.drop(columns=['month'], inplace=True)
df_clean.drop(columns=['pdays'], inplace=True)
df_clean.head()
```

```python
# check the remaining of columns

df_clean.columns
```

```python
# check the data type

df_clean.info()
```

- Note that many of the columns are not in numerical data type, therefore not suitable yet for building ML model.

- We need to encode the columns that are not in numerical datatype.

```python
# make a list of columns to be encoded from string to integer
encoded_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan',
               'contact', 'poutcome', 'y']

# run the encoder
encoder = LabelEncoder()

for col in encoded_cols:
    df_clean[col] = encoder.fit_transform(df_clean[col])
    print(col, encoder.classes_)
```

```python
# now, check again the data type. they should all have int or float datatype.

df_clean.info()
```

# II. Building Machine Learning Model

## II. 1. Data Preparation

```python
# first, separate features and target

features = df_clean[['age', 'job', 'marital', 'education', 'default', 'housing', 'loan',
       'contact', 'campaign', 'previous',
       'poutcome', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx',
       'euribor3m', 'nr.employed']]
target = df_clean['y']

# check the shape to ensure it's correct

print(features.shape, target.shape)
```

```python
# split training data and test data

x_train, x_test, y_train, y_test = train_test_split(features, target, random_state=0, test_size=0.2)

print(x_train.shape, x_test.shape)
```

```python
# we will also later need to visualize the confusion matrix
# we will keep the function over here to be used in each model later

def visualizeCM(cm):    
    nolabels = len(cm)
    fig, ax = plt.subplots(figsize=(nolabels, nolabels))
    ax.imshow(cm)
    ax.grid(False)
    ax.set_xlabel('Predicted outputs', color='black')
    ax.set_ylabel('Actual outputs', color='black')
    ax.xaxis.set(ticks=range(nolabels))
    ax.yaxis.set(ticks=range(nolabels))
    ax.set_ylim(nolabels-0.5, -0.5)
    for i in range(nolabels):
        for j in range(nolabels):
            ax.text(j, i, cm[i, j], ha='center', va='center', color='black')
    plt.show()
```

## II. 2. Model Building

## II.2.1. **Random Forest**

- We will start with Random Forest algorithm because it has a function to see the importance of each feature.
- From here, hopefully we'll be able to see which feature is not relevant therefore we can exclude that feature for other ML models.

```python
# build and train model
model_rf = RandomForestClassifier(random_state=0).fit(x_train, y_train)
print(model_rf)

# evaluate model
y_pred = model_rf.predict(x_test)

# check the result report and the visualization matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
visualizeCM(confusion_matrix(y_test, y_pred))
```

```python
# we will now try to use max_depth = 3
# notice that the accuracy increases slightly to 89%
# max_depth = 2 was also previously run but not displayed in this notebook, it gives similar result to the model above

# create and train model
model_rf3_test = RandomForestClassifier(max_depth=3).fit(x_train, y_train)
print(model_rf3_test)

# evaluate model
y_pred = model_rf3_test.predict(x_test)

# check the result report and the visualization matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
visualizeCM(confusion_matrix(y_test, y_pred))
```

### Using training dataset (to check any possibility of overfitting)

- Note that there is similar result of accuracy and F1 score between the models that used training dataset and testing dataset, when using max_depth=3.

## **II. 2. 4. K-Nearest Neighbor**

- K-NN algorithm assumes the similarity between the new case/data and available cases and put the new case into the category that is most similar to the available categories. (Source: https://www.javatpoint.com/k-nearest-neighbor-algorithm-for-machine-learning)

```python
# first, we will look at how many 'K' or iteration required until the model is stable
# from the graph, we can see that it starts becoming stable at approximately 8

def knnTableTrainTest(maxK, x_train, y_train, x_test, y_test):
    list_score = []
    for n in range(1, maxK):
        model_knn = KNeighborsClassifier(n_neighbors = n).fit(x_train, y_train)
        score = model_knn.score(x_test, y_test)
        list_score.append(score)

    plt.plot(range(1,maxK), list_score, marker="o")
    plt.xlabel("K neighbor numbers")
    plt.ylabel("Accuracy Score")
    plt.title("KNN score vs K table")
    
knnTableTrainTest(15, x_train, y_train, x_test, y_test)
```

### Tuning: Using n_neighbors

```python
# create and train model
model_knn = KNeighborsClassifier(n_neighbors=8).fit(x_train, y_train)
print(model_knn)

# evaluate model
y_pred = model_knn.predict(x_test)

# check the result report and the visualization matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
visualizeCM(confusion_matrix(y_test, y_pred))
```

### Using training dataset (to check any possibility of overfitting)

- To ensure that the model is not overfitted, we will check with the train dataset as well.
- If the accuracy between using the training dataset vs. testing dataset is not very contrasting, the model is assumed to be acceptable.

```python
# create and train model
model_knn_train = KNeighborsClassifier(n_neighbors=8).fit(x_train, y_train)
print(model_knn_train)

# evaluate model using training dataset
y_pred = model_knn_train.predict(x_train)

# check the result report and the visualization matrix
print(classification_report(y_train, y_pred))
print(confusion_matrix(y_train, y_pred))
visualizeCM(confusion_matrix(y_train, y_pred))
```

### Analysis of K-Nearest Neighbors model:

## II. 2. 6. Logistic Regression

- Logistic regression is a process of modeling the probability of a discrete outcome given an input variable. The most common logistic regression models a binary outcome; something that can take two values such as true/false, yes/no, and so on. (Source: Edgar & Manz, 2017 (https://www.sciencedirect.com/science/article/pii/B9780128053492000042))
- There are several solvers available, however we will use liblinear and lbfgs for reasons that will be elaborated below.

```python
# create and train model
model_lr = LogisticRegression(solver='liblinear', random_state=0).fit(x_train, y_train)
print(model_lr)

# evaluate model
y_pred = model_lr.predict(x_test)

# check the result report and the visualization matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
visualizeCM(confusion_matrix(y_test, y_pred))
```

### Tuning: Using 'lbfgs' and 'max_iter'

- Solver 'liblinear' did not result in any error, however it is usually used for small dataset.
- We will use 'lbfgs' which tends to fit for all, while also tune the maximum iteration.

# IV. Conclusion

- We used 6 types of supervised machine learning algorithms to create 6 different ML classification models.
- Based on its accuracy, the order in which the accuracy is the highest to lowest are: Random Forest model, Decision Tree model and Logistic Regression model (both are similar), K-Nearest Neighbors and Support Vector Machine models (both are similar), and Naive Bayes model.
- After making sure the models do not overfit, Random Forest was found to yield the best results compared to other algorithms. The possible reason is summarized from https://towardsdatascience.com/understanding-random-forest-58381e0602d2:

_"The low correlation between models is the key. Just like how investments with low correlations (like stocks and bonds) come together to form a portfolio that is greater than the sum of its parts, uncorrelated models can produce ensemble predictions that are more accurate than any of the individual predictions. The reason for this wonderful effect is that the trees protect each other from their individual errors (as long as they don’t constantly all err in the same direction). While some trees may be wrong, many other trees will be right, so as a group the trees are able to move in the correct direction."_

- Furthermore, using Random Forest algorithm first resulted in a finding that several features can be discarded as they affect very little to none to the target, such as 'job', 'marital', 'education', 'default', 'housing', 'loan', and 'campaign' 

- Therefore, Random Forest is the algorithm of choice in this assignment because not only that it offers a more robust analysis as a result of multiple of decision trees that are put to a vote, but also proven in the accuracy level, and theoretically suitable for large datasets such as this dataset. From data visualization using bar charts and pie charts, we can also easily see why the algorithm gives higher importance level to certain feature compared to the other, therefore easier to understand.

- Supporting the feature importance analysis of Random Forest algorithm, there seems to be almost no relationship between the discarded features of each population with the decision of whether or not they are subscribed to the term deposit as shown by their distribution in each population which are relatively very similar across the two populations. The only difference is that there are higher proportion of blue-collared workers who do not subscribe, therefore indicating that while not all social and economic factors are at play, it may be considerable for certain demographic only such as the blue-collared workers.
- The quality of each campaign is also possibly a lot more significant rather than the quantity of the campaigns or contacts. Other factors that may affect the decision to subscribe or not may include the trust gained with the bank, which can be gained through frequent contacts and ability of the bank staff to maintain long call durations.
